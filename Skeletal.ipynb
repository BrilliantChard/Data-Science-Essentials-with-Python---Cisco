```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bird_data = pd.read_csv('bird.csv')
bird_data.head()
```

```python
counts = bird_data['neck_vertebrae'].value_counts()
counts = counts.sort_index()
```

```python
plt.figure(figsize=(8,6))
plt.bar(counts.index, counts.values)

plt.xlabel("Number of Neck Vertebrae")
plt.ylabel("Count of Birds")
plt.title("Distribution of Neck Vertebrae in Birds")
plt.show()
```

```python
bird_counts = bird_data['neck_vertebrae'].value_counts()
bird_counts = bird_counts.sort_index()
bird_counts.plot.bar()
```
